## DATA LOADING

#### 모든 구현 디테일은 논문과 그대로 진행하였습니다.
#### 성능을 끌어 올리기 위하여 *Bag of Tricks for Image Classification with Convolutional Neural Networks (cvpr 2019)*에 소개된 프로세스를 다수 적용하였습니다.

```python
import torch
from torch.autograd import Variable
from torchvision import datasets, transforms

from model import CapsNet
from smallNorb import SmallNORB
from tools import *

from tqdm import tqdm

test_transform = transforms.Compose([
    transforms.Resize(48),
    transforms.CenterCrop(32),
    transforms.ToTensor(),
    transforms.Normalize((0.,), (0.3081,))
    ])

test_dataset = SmallNORB('./datasets/smallNORB/', train=False, transform=test_transform)
#test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=128, shuffle=True, num_workers=2)
```

```python
import os
path = './TestBinImg'
path2 = './TestBinLabel'
```

```python
for name, i in enumerate(test_dataset):
    img, label = i
    save_img_name = os.path.join(path, str(name)+'.bin')
    save_label_name = os.path.join(path2, str(name)+'.bin')
    with open(save_img_name, 'wb') as f:
        f.write(img.numpy())
    with open(save_label_name, 'wb') as f:
        f.write(label.numpy())
    print(name/len(test_dataset))
```

## MODEL LOADING

#### 본 논문과는 다르게 batch norm을 적용하였습니다. batchnorm False로 설정하시면 적용을 안하실 수 있습니다

```python
capsnet = CapsNet(reconstruction_type='FC',
        imsize=32,
        num_classes=5,
        routing_iterations = 3,
        primary_caps_gridsize=8,
        num_primary_capsules=32,
        batchnorm=True,
        loss = 'L2',
        leaky_routing=False)

capsnet = capsnet.cuda()
checkpoint = torch.load('./checkpoint/96.3.pt') 
capsnet.load_state_dict(checkpoint['net'])
```

## TEST

```python
def test(epoch):
    global best_acc
    capsnet.eval()
    test_correct = 0
    total = 0

    for batch_id, (data, target) in tqdm(list(enumerate(test_loader)), ascii=True, desc="Test {:3d}".format(epoch)):
        data, target = Variable(data), Variable(target) # (batch size, channels, hight, wight)
        
        data, target = data.cuda(), target.cuda()
        target = torch.eye(5).cuda().index_select(dim=0, index=target)

        capsule_output, reconstructions, predictions = capsnet(data) # (predictions = class output)
        data = denormalize(data)
        loss, rec_loss, marg_loss = capsnet.loss(data, target, capsule_output, reconstructions, 0.0005)

        test_correct += (target.max(dim=1)[1] == predictions.max(dim=1)[1]).sum().item()
        total += target.size(0)
    acc = 100.*test_correct/total
    print("acc = {}%".format(test_correct/total))
```
